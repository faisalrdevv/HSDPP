```python
import pandas as pd
import numpy as numpy
import neattext.functions as nfx
```

```python
clear_tweets = []
df = pd.read_csv('/workspaces/HSDPP/Twitter SN Dataset.csv', index_col=False)
df.head()
```

```python
tlist = []
for x in range(len(df)):
    tw = df.tweet[x]
    s1 = nfx.clean_text(tw,urls=True,emails=True,numbers=True)
    s2 = nfx.remove_hashtags(s1)
    s3 = nfx.remove_puncts(s2)
    tlist.append(s3)
```

```python
df['clean_tweet']=tlist
```
